# Metodologia:

- Divide & Conquer: Divis√£o do intervalo em subintervalos e c√°lculo das m√©dias.

# 1 simula o crescimento do PIB com o m√©todo de Monte Carlo, usando 10.000 rodadas e distribui√ß√£o uniforme entre 2% e 4%:


```python
import numpy as np

# Fun√ß√£o para simular o crescimento do PIB com Monte Carlo
def monte_carlo_pib(n=10000):
    # Gera n valores aleat√≥rios com distribui√ß√£o uniforme entre 2% e 4%
    simulacoes = np.random.uniform(low=0.02, high=0.04, size=n)

    # Calcula a m√©dia e o desvio padr√£o das simula√ß√µes
    media = np.mean(simulacoes)
    desvio = np.std(simulacoes)

    # Retorna os resultados
    return simulacoes, media, desvio

# Executa a simula√ß√£o
dados_pib, media_pib, desvio_pib = monte_carlo_pib()

# Exibe os resultados
print(f"M√©dia estimada do crescimento do PIB: {media_pib:.4f} ou {media_pib*100:.2f}%")
print(f"Desvio padr√£o estimado: {desvio_pib:.4f} ou {desvio_pib*100:.2f}%")
```

# 2 usando o m√©todo de Divide & Conquer dividindo o intervalo [2%, 4%] em  subintervalos e calculando as m√©dias.


```python
import numpy as np

# Fun√ß√£o para estimar o crescimento do PIB usando Divide & Conquer
def divide_conquer_pib(intervalo=(0.02, 0.04), partes=10):
    # Divide o intervalo [2%, 4%] em 'partes' subintervalos iguais
    subintervalos = np.linspace(intervalo[0], intervalo[1], partes + 1)

    # Calcula a m√©dia de cada subintervalo
    medias = [(subintervalos[i] + subintervalos[i+1]) / 2 for i in range(partes)]

    # Calcula a m√©dia geral e o desvio padr√£o das m√©dias
    media_final = np.mean(medias)
    desvio_final = np.std(medias)

    # Retorna os valores
    return medias, media_final, desvio_final

# Executa o m√©todo Divide & Conquer
dados_dc, media_dc, desvio_dc = divide_conquer_pib()

# Exibe os resultados
print(f"M√©dia estimada (Divide & Conquer): {media_dc:.4f} ou {media_dc*100:.2f}%")
print(f"Desvio padr√£o estimado: {desvio_dc:.4f} ou {desvio_dc*100:.2f}%")
```

# 4 Estrutura com Decorators e Boas Pr√°ticas

```python
import numpy as np
import matplotlib.pyplot as plt
import time

# ‚úÖ Decorator para medir tempo de execu√ß√£o
def medir_tempo(func):
    def wrapper(*args, **kwargs):
        inicio = time.time()
        resultado = func(*args, **kwargs)
        fim = time.time()
        print(f"‚è± Tempo de execu√ß√£o de '{func.__name__}': {fim - inicio:.4f} segundos")
        return resultado
    return wrapper

# ‚úÖ Fun√ß√£o Monte Carlo com 10.000 simula√ß√µes
@medir_tempo
def simular_pib_monte_carlo(n=10000, minimo=0.02, maximo=0.04):
    """
    Simula o crescimento do PIB com distribui√ß√£o uniforme.
    Retorna os dados simulados, m√©dia e desvio padr√£o.
    """
    simulacoes = np.random.uniform(low=minimo, high=maximo, size=n)
    media = np.mean(simulacoes)
    desvio = np.std(simulacoes)
    return simulacoes, media, desvio

# ‚úÖ Fun√ß√£o Divide & Conquer com divis√£o em subintervalos
@medir_tempo
def estimar_pib_divide_conquer(intervalo=(0.02, 0.04), partes=10):
    """
    Divide o intervalo em subintervalos e calcula a m√©dia de cada.
    Retorna os valores m√©dios, m√©dia geral e desvio padr√£o.
    """
    subintervalos = np.linspace(intervalo[0], intervalo[1], partes + 1)
    medias = [(subintervalos[i] + subintervalos[i+1]) / 2 for i in range(partes)]
    media_final = np.mean(medias)
    desvio_final = np.std(medias)
    return medias, media_final, desvio_final
```

Esse c√≥digo gera 50.000 cen√°rios de exporta√ß√µes e importa√ß√µes com base nas distribui√ß√µes normais indicadas, calcula a balan√ßa de pagamentos e exibe a m√©dia e o desvio padr√£o dos resultados.

# 2 Resolvendo com Dynamic Programming (Bottom ‚Äì Up), dividindo as distribui√ß√µes em quartis e combinando cen√°rios.

```python
import numpy as np

# Fun√ß√£o para simular a balan√ßa de pagamentos com Monte Carlo
def monte_carlo_balanca(n=50000):
    # Exporta√ß√µes: distribui√ß√£o normal com m√©dia 100 e desvio padr√£o 10
    exportacoes = np.random.normal(loc=100, scale=10, size=n)

    # Importa√ß√µes: distribui√ß√£o normal com m√©dia 90 e desvio padr√£o 15
    importacoes = np.random.normal(loc=90, scale=15, size=n)

    # Balan√ßa de pagamentos = exporta√ß√µes - importa√ß√µes
    balanca = exportacoes - importacoes

    # Estat√≠sticas da simula√ß√£o
    media = np.mean(balanca)
    desvio = np.std(balanca)

    # Retorna os dados simulados e estat√≠sticas
    return balanca, media, desvio

# Executando a simula√ß√£o
balanca_mc, media_mc, desvio_mc = monte_carlo_balanca()

# Exibindo os resultados
print(f"M√©dia da balan√ßa de pagamentos: {media_mc:.2f} bilh√µes USD")
print(f"Desvio padr√£o: {desvio_mc:.2f} bilh√µes USD")
```

```python
import numpy as np

# Fun√ß√£o para estimar a balan√ßa de pagamentos usando quartis e combina√ß√£o de cen√°rios
def dp_balanca_quartis():
    # Gerando uma amostra grande para estimar os quartis com precis√£o
    exportacoes_amostra = np.random.normal(loc=100, scale=10, size=100000)
    importacoes_amostra = np.random.normal(loc=90, scale=15, size=100000)

    # Calculando os quartis (25%, 50%, 75%) e adicionando extremos aproximados
    export_quartis = np.percentile(exportacoes_amostra, [25, 50, 75])
    export_quartis = [np.min(exportacoes_amostra)] + list(export_quartis) + [np.max(exportacoes_amostra)]

    import_quartis = np.percentile(importacoes_amostra, [25, 50, 75])
    import_quartis = [np.min(importacoes_amostra)] + list(import_quartis) + [np.max(importacoes_amostra)]

    # Combinando todos os cen√°rios poss√≠veis entre os quartis
    balanca_aprox = []
    for e in export_quartis:
        for i in import_quartis:
            saldo = e - i
            balanca_aprox.append(saldo)

    # Convertendo para array NumPy e calculando estat√≠sticas
    balanca_aprox = np.array(balanca_aprox)
    media = np.mean(balanca_aprox)
    desvio = np.std(balanca_aprox)

    return balanca_aprox, media, desvio

# Executando o m√©todo din√¢mico
balanca_dp, media_dp, desvio_dp = dp_balanca_quartis()

# Exibindo os resultados
print(f"M√©dia da balan√ßa (Din√¢mico): {media_dp:.2f} bilh√µes USD")
print(f"Desvio padr√£o (Din√¢mico): {desvio_dp:.2f} bilh√µes USD")
```

- Simula cen√°rios discretos com base nos quartis das distribui√ß√µes de exporta√ß√µes e importa√ß√µes.
- Calcula todas as combina√ß√µes poss√≠veis entre os quartis.
- Estima a balan√ßa de pagamentos com base nessas combina√ß√µes.
- Retorna a m√©dia e o desvio padr√£o da distribui√ß√£o aproximada.

Simular e estimar o PIB ap√≥s 10 anos e o saldo acumulado da balan√ßa de pagamentos de um pa√≠s, considerando incertezas nas taxas de crescimento, exporta√ß√µes e importa√ß√µes. O exerc√≠cio aplica dois m√©todos distintos: Monte Carlo e Brute Force (parti√ß√£o temporal), permitindo a compara√ß√£o entre uma abordagem estat√≠stica estoc√°stica e uma segmentada determin√≠stica.

- Brute Force: Parti√ß√£o do per√≠odo em dois blocos (anos 1‚Äì5 e 6‚Äì10), com recombina√ß√£o dos resultados.

# 1 Usar Monte Carlo (100.000 simula√ß√µes) para estimar

```python
import numpy as np
import matplotlib.pyplot as plt
import time

# ‚úÖ Decorator para medir tempo de execu√ß√£o
def medir_tempo(func):
    def wrapper(*args, **kwargs):
        inicio = time.time()
        resultado = func(*args, **kwargs)
        fim = time.time()
        print(f"‚è± Tempo de execu√ß√£o de '{func.__name__}': {fim - inicio:.4f} segundos")
        return resultado
    return wrapper

# ‚úÖ Fun√ß√£o Monte Carlo para 100.000 simula√ß√µes ao longo de 10 anos
@medir_tempo
def simular_cenarios_longo_prazo(n=100000, anos=10, pib_inicial=2_000_000_000_000):
    """
    Simula o crescimento do PIB e a balan√ßa de pagamentos acumulada ao longo de 10 anos.
    PIB em trilh√µes, exporta√ß√µes/importa√ß√µes em bilh√µes.
    """
    # Simula taxas de crescimento anual do PIB (normal: m√©dia 3%, desvio 1%)
    taxas_pib = np.random.normal(loc=0.03, scale=0.01, size=(n, anos))

    # Calcula o PIB final ap√≥s 10 anos para cada simula√ß√£o
    pib_final = pib_inicial * np.prod(1 + taxas_pib, axis=1)

    # Simula exporta√ß√µes e importa√ß√µes anuais (normal como no exerc√≠cio 2)
    exportacoes = np.random.normal(loc=100, scale=10, size=(n, anos))
    importacoes = np.random.normal(loc=90, scale=15, size=(n, anos))

    # Calcula balan√ßa de pagamentos acumulada ao longo dos 10 anos
    balanca_total = np.sum(exportacoes - importacoes, axis=1)

    # Estat√≠sticas
    media_pib = np.mean(pib_final)
    desvio_pib = np.std(pib_final)

    media_balanca = np.mean(balanca_total)
    desvio_balanca = np.std(balanca_total)

    return pib_final, balanca_total, media_pib, desvio_pib, media_balanca, desvio_balanca
```

```python
pib_final, balanca_total, media_pib, desvio_pib, media_balanca, desvio_balanca = simular_cenarios_longo_prazo()

# Exibindo os resultados
print(f"\nüìà PIB ap√≥s 10 anos (m√©dia): {media_pib / 1e12:.2f} trilh√µes USD")
print(f"Desvio padr√£o do PIB: {desvio_pib / 1e12:.2f} trilh√µes USD")

print(f"\nüí∞ Balan√ßa acumulada (m√©dia): {media_balanca:.2f} bilh√µes USD")
print(f"Desvio padr√£o da balan√ßa: {desvio_balanca:.2f} bilh√µes USD")
```

# usamos o m√©todo de Monte Carlo para projetar o comportamento econ√¥mico de um pa√≠s ao longo de 10 anos. A cada simula√ß√£o:

- O PIB inicial de 2 trilh√µes cresce ano a ano com uma taxa aleat√≥ria (normal, m√©dia 3%, desvio 1%).
- As exporta√ß√µes e importa√ß√µes tamb√©m variam anualmente (normal, como no exerc√≠cio 2).
- Calculamos o PIB final ap√≥s 10 anos e a balan√ßa de pagamentos acumulada somando os saldos anuais.
Repetimos esse processo 100.000 vezes para obter uma distribui√ß√£o estat√≠stica dos poss√≠veis cen√°rios econ√¥micos futuros.


# 2  Resolvendo com Brute Force, particionando o per√≠odo em metades
# (anos 1‚Äì5, anos 6‚Äì10), resolvendo cada parte e recombinando resultados.

```python
import numpy as np
import time

# ‚úÖ Decorator para medir tempo
def medir_tempo(func):
    def wrapper(*args, **kwargs):
        inicio = time.time()
        resultado = func(*args, **kwargs)
        fim = time.time()
        print(f"‚è± Tempo de execu√ß√£o de '{func.__name__}': {fim - inicio:.4f} segundos")
        return resultado
    return wrapper

# ‚úÖ Fun√ß√£o Brute Force com divis√£o em 2 blocos de 5 anos
@medir_tempo
def simular_brute_force(n=100000, pib_inicial=2_000_000_000_000):
    """
    Simula o crescimento do PIB e a balan√ßa de pagamentos acumulada em duas etapas:
    anos 1‚Äì5 e anos 6‚Äì10.
    """
    # Simula taxas de crescimento para cada bloco
    taxas_1 = np.random.normal(0.03, 0.01, size=(n, 5))
    taxas_2 = np.random.normal(0.03, 0.01, size=(n, 5))

    # PIB ap√≥s 5 anos
    pib_5 = pib_inicial * np.prod(1 + taxas_1, axis=1)

    # PIB ap√≥s 10 anos (recombinando)
    pib_10 = pib_5 * np.prod(1 + taxas_2, axis=1)

    # Exporta√ß√µes e importa√ß√µes em cada bloco
    exp_1 = np.random.normal(100, 10, size=(n, 5))
    imp_1 = np.random.normal(90, 15, size=(n, 5))
    exp_2 = np.random.normal(100, 10, size=(n, 5))
    imp_2 = np.random.normal(90, 15, size=(n, 5))

    # Balan√ßa acumulada por bloco
    balanca_1 = np.sum(exp_1 - imp_1, axis=1)
    balanca_2 = np.sum(exp_2 - imp_2, axis=1)

    # Balan√ßa total recombinada
    balanca_total = balanca_1 + balanca_2

    # Estat√≠sticas
    media_pib = np.mean(pib_10)
    desvio_pib = np.std(pib_10)
    media_balanca = np.mean(balanca_total)
    desvio_balanca = np.std(balanca_total)

    return pib_10, balanca_total, media_pib, desvio_pib, media_balanca, desvio_balanca
```

```python
pib_final_bf, balanca_bf, media_pib_bf, desvio_pib_bf, media_balanca_bf, desvio_balanca_bf = simular_brute_force()

print(f"\nüìà PIB ap√≥s 10 anos (Brute Force): {media_pib_bf / 1e12:.2f} trilh√µes USD")
print(f"Desvio padr√£o do PIB: {desvio_pib_bf / 1e12:.2f} trilh√µes USD")

print(f"\nüí∞ Balan√ßa acumulada (Brute Force): {media_balanca_bf:.2f} bilh√µes USD")
print(f"Desvio padr√£o da balan√ßa: {desvio_balanca_bf:.2f} bilh√µes USD")
```

- A abordagem Brute Force divide o problema em partes menores e resolve cada uma separadamente.
- Isso permite analisar blocos de tempo distintos, o que pode ser √∫til para pol√≠ticas econ√¥micas em fases.
- Os resultados s√£o recombinados para obter o PIB final e a balan√ßa acumulada
